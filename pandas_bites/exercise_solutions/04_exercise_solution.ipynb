```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
planets = pd.read_csv('./planets.csv')

planets
```

## List all the planets which have a greater mass than Earth.

```python
# #iloc() and position 2 in the list of planets, then the column name
earth_mass = planets.iloc[2]['Mass (10^24kg)']

massive_list = planets.loc[planets['Mass (10^24kg)'] > earth_mass]

massive_list

```

```python
ring_systems = planets.loc[planets['Ring System?'] == 'Yes']

ring_systems
```

## Find the maximum surface gravity value for a planet in our solar system. Which planet is it?

```python
max_gravity = planets['Surface Gravity(m/s^2)'].max()

print(max_gravity )

max_gravity_planet = planets.loc[planets['Surface Gravity(m/s^2)'] == max_gravity]

max_gravity_planet
```

## Find the total Mass of all the planets in our solar system. 

```python
planetary_mass = planets['Mass (10^24kg)'].sum()

print(planetary_mass)

sun_mass = 1989000

print(f"The Sun in our solar system is 1.989 × 10^30 kg, which is 6 orders of magnitude more than our metric for planetary mass, but whereas the combined mass of the planet of our solar system is {planetary_mass}, this pales in comparison to the mass of the Sun, which is {sun_mass}.")

```

```python
print(f"You could fit the mass of all the planets {round(sun_mass / planetary_mass)} times into the Sun")
```

## For an extra stretch goal, revisit matplotlib, and chart the planets distance from the Sun against their Diameter.

```python
x = planets['Distance from Sun (10^6 km)']
y = planets['Diameter (km)']

plt.plot(x, y)
plt.xlabel('Distance from Sun (10^6 km)')
plt.ylabel('Diameter (km)')
plt.show()
```
