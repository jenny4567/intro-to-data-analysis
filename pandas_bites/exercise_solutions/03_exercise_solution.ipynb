```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hap18 = pd.read_csv("./happiness_data/2018.csv")

hap19 = pd.read_csv("./happiness_data/2019.csv")
```

```python
hap18_and_19_merged = hap18.join(hap19, how="inner", lsuffix="_2018", rsuffix="_2019")

hap18_and_19_merged.head()

# This joined table is too wide to display all the relevent information
# We can select specific columns
```

```python

rank_country_score_merge = hap18[['Overall rank','Country or region', 'Score']].join(hap19[['Overall rank','Country or region', 'Score']], how="inner", lsuffix="_2018", rsuffix="_2019")

rank_country_score_merge.head()

# This still looks a little unclear as the ranks are in the same order, but the countries are different
# Lets explore a similar take but with #merge()
```
