Let's look at the World Happiness Data once again, and use it to plot some charts and infer some results.

Does the GDP data for the United Kingdom seem to have an influence on the overall standings in Happiness Rank?

Chart the yearly Happiness Rank against the yearly GDP, and visualise your conclusion.

```python
import pandas as pd
import matplotlib.pyplot as plt

hap15 = pd.read_csv("./happiness_data/2015.csv")
hap16 = pd.read_csv("./happiness_data/2016.csv")
hap17 = pd.read_csv("./happiness_data/2017_missing.csv")
hap18 = pd.read_csv("./happiness_data/2018.csv")
hap19 = pd.read_csv("./happiness_data/2019.csv")

# Check the varying column names 

hap16.columns
```

## What Statistical Data Types are `Happiness Rank` and `GDP per capita`? Explain in your notebook.


```python
hap19['GDP per capita'].head()
```

`GDP per capita` is an example of Continuous data. It can take any value between a range, and is not a count, unlike...

```python
hap19['Overall rank'].head()
```

`Overall rank` or Happiness Rank, is a an example of Discrete data, as it is a count. It cannot be a float - a country cannot be Rank 1.4, or 2.3. They are either first, or second etc.

```python
y1 = [
hap15.loc[hap15['Country']=='United Kingdom']['Happiness Rank'],
hap16.loc[hap16['Country']=='United Kingdom']['Happiness Rank'],
hap17.loc[hap17['Country']=='United Kingdom']['Happiness.Rank'],
hap18.loc[hap18['Country or region']=='United Kingdom']['Overall rank'],
hap19.loc[hap19['Country or region']=='United Kingdom']['Overall rank']]

for i in y1:
    print(i)
```

```python
y2 = [
(hap15.loc[hap15['Country']=='United Kingdom']['Economy (GDP per Capita)']),
hap16.loc[hap16['Country']=='United Kingdom']['Economy (GDP per Capita)'],
hap17.loc[hap17['Country']=='United Kingdom']['Economy..GDP.per.Capita.'],
hap18.loc[hap18['Country or region']=='United Kingdom']['GDP per capita'],
hap19.loc[hap19['Country or region']=='United Kingdom']['GDP per capita']]

for i in y2:
    print(i)
```

```python
x = [
    '2015', '2016', '2017', '2018', '2019'
]
```

In two plots...

```python
plt.plot(x, y1)
plt.xlabel('Year')
plt.ylabel('Happiness Rank')
```

```python
plt.plot(x, y2)
plt.xlabel('Year')
plt.ylabel('GDP')

plt.show()
```

### As a stretch goal, can you plot both the GDP axis and the Happiness axis on the same chart to show the similarities?

in one plot

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Happiness Rank', color='r')
ax2.set_ylabel('GDP', color='b')
```
