# Clean and filter data with pandas - Temperature

```python
import pandas as pd
import numpy as np

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# We are using a random number generator similar to the #random library in python
# This is a Numpy Random Number Generator that can produce numbers ina  certain shape
# In this example 5 working days worth, over 3 weeks times 6 for some chilly weather (in Celsius)!

temperature_df = pd.DataFrame((np.random.randn(5, 3)*6), index=days,columns=['week one', 'week two', 'week three'])

print(temperature_df) 
```

```python
# What about the weekend?

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

temperature_df = temperature_df.reindex(days)

print(temperature_df)
```

```python
temperature_df.isnull()
# Checks for null values

temperature_df.notnull()
# Checks for non-null (apparent) values
```

```python

for week in temperature_df:
    week_mean = temperature_df[week].mean()
    temperature_df[week] = temperature_df[week].fillna(week_mean)

temperature_df
```

```python
tree_df = pd.DataFrame({'tree_ages': [2, 5, 7, 8, 14, 16, 17, 25, 32, 33],
'tree_heights': [1.8, 3.6, 4.1, 3.9, 7.2, 7.5, 8.1, 11, 15, 14.1]})
tree_df.corr()
```

Can you calculate the correlation coefficient for our tree data with just a calculator? Google the process.



```python
mean_tree_age = mean(tree_df.tree_ages)
mean_tree_height = mean(tree_df.tree_heights)

top_of_eq = 0
for i in range(0,len(tree_df.tree_ages)):
    element = (tree_df.tree_ages[i] - mean_tree_age) * (tree_df.tree_heights[i] - mean_tree_height)
    top_of_eq += element

bottom_of_eq = np.sqrt(sum([(x - mean_tree_age)**2 for x in tree_df.tree_ages])) * np.sqrt(sum([(x - mean_tree_height)**2 for x in tree_df.tree_heights]))

print('The calculated correlation coefficient is: ',top_of_eq / bottom_of_eq)
print('The real correlation coefficient is: ', tree_df.corr())
```

```python
anscombe_quartet = pd.read_csv('/Users/MakersAdmin/Documents/my_python_code/intro-to-data-analysis/data/smaller-datasets/Anscombe_quartet_data.csv')
anscombe_quartet

for column in anscombe_quartet.columns.values.tolist():
    print(column, " mean is ", mean(anscombe_quartet[column]))

print('The correlation between the first pair of values is: ', anscombe_quartet['x1'].corr(anscombe_quartet['y1']))
print('The correlation between the second pair of values is: ', anscombe_quartet['x2'].corr(anscombe_quartet['y2']))
print('The correlation between the third pair of values is: ', anscombe_quartet['x3'].corr(anscombe_quartet['y3']))
print('The correlation between the fourth pair of values is: ', anscombe_quartet['x4'].corr(anscombe_quartet['y4']))
```

Make your own data set where x always takes the same value (let's say 5) and the computer chooses random values for y. What do you get as the correlation coefficient?

Surely the line of best fit is x = 5

Can you get that out of numpy?

```python
from random import randint

x = 10 * [5]
y = [randint(0,10) for x in range(0,10)]
df=pd.DataFrame({
    'x':x,
    'y':y
})
print(df)
df.plot.scatter('x','y')
print('x and y correlation is ', df.x.corr(df.y))
```

Previously, we tried make data with low correlation by picking random numbers. As done in this notebook. We didn't quite get zero with random numbers that time. Well, you may have heard that computers aren't actually able to pick random numbers -- they are just pseudo-random numbers. So let's use some dice. A red one and a green one. Here is my collected data, after two rolls of each die:
Plot a scattergraph, with a line of best fit and find the correlation coefficient. Surely it will be very close to zero, after all this is genuinely random.

```python
data = pd.DataFrame({
   "scores on red": [3, 5],
   "scores on green": [2, 1]
})
data.plot.scatter("scores on red","scores on green")
data.corr()
```

So two rolls of each die is not going to be enough -- our sample is too small. Try adding more rolls to your data and see what you get. (You can simulate rolls with a computer, if you like). 10 rolls of each die would certainly be more convincing.

```python

data = pd.DataFrame({
   "scores on red": [randint(1,7) for x in range(0,10)],
   "scores on green": [randint(1,7) for x in range(0,10)]
})
data.plot.scatter("scores on red","scores on green")
data.corr()
```
